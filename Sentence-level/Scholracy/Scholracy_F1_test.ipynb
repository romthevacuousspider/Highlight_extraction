```python
#Import test set (same for Llama3 and Open AI API)
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Final Project/Field Project/Highlighted Papers/test_dataframe_scholracy - test_dataframe_scholracy.csv')
```

```python
df.head(2)
```

```python
# getting F1-score
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, f1_score

def get_performance_metrics(df):
    y_test = df.Highlight
    y_pred = df.Predict

    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    print("Accuracy Score:", accuracy_score(y_test, y_pred))
    print("F-1 Score:", f1_score(y_test, y_pred, average='weighted'))

```

```python
get_performance_metrics(df)
```

```python
grouped = df.groupby('Field')

# Filter by 'Field'
field_datasets = {field: group_df.reset_index(drop=True) for field, group_df in grouped}
```

```python
for field, dataset in field_datasets.items():
    print(f"Performance metrics for {field}:")
    get_performance_metrics(dataset)
    print("\n" + "-"*50 + "\n")
```
